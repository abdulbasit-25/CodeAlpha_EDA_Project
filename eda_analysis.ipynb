```python
# =========================================
# Exploratory Data Analysis (EDA)
# Employee Dataset - Google Colab Version
# =========================================

# -----------------------------------------
# 1. Import Required Libraries
# -----------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Improve plot visuals
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

# -----------------------------------------
# 2. Upload Dataset from Computer
# -----------------------------------------

from google.colab import files
uploaded = files.upload()

# After upload, load the dataset
df = pd.read_csv("employee_dummy_dataset.csv")

print("Dataset Loaded Successfully!")

# -----------------------------------------
# 3. Display Basic Information
# -----------------------------------------

print("\nFirst 5 Rows of Dataset")
print(df.head())

print("\nDataset Shape (Rows, Columns)")
print(df.shape)

print("\nColumn Names")
print(df.columns)

print("\nDataset Information")
df.info()

# -----------------------------------------
# 4. Statistical Summary
# -----------------------------------------

print("\nStatistical Summary")
print(df.describe())

# -----------------------------------------
# 5. Check Missing Values
# -----------------------------------------

print("\nMissing Values per Column")
print(df.isnull().sum())

# -----------------------------------------
# 6. Unique Values
# -----------------------------------------

print("\nUnique Departments")
print(df["Department"].unique())

print("\nUnique Performance Ratings")
print(df["Performance"].unique())

# -----------------------------------------
# 7. Age Distribution
# -----------------------------------------

plt.figure()
sns.histplot(df["Age"], bins=10, kde=True)
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# -----------------------------------------
# 8. Salary Distribution
# -----------------------------------------

plt.figure()
sns.histplot(df["Salary"], bins=10, kde=True)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()

# -----------------------------------------
# 9. Employees per Department
# -----------------------------------------

plt.figure()
sns.countplot(x="Department", data=df)
plt.title("Employees per Department")
plt.xticks(rotation=30)
plt.show()

# -----------------------------------------
# 10. Gender Distribution
# -----------------------------------------

plt.figure()
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()

# -----------------------------------------
# 11. Performance Rating Distribution
# -----------------------------------------

plt.figure()
sns.countplot(x="Performance", data=df)
plt.title("Performance Rating Distribution")
plt.show()

# -----------------------------------------
# 12. Experience vs Salary
# -----------------------------------------

plt.figure()
sns.scatterplot(x="YearsExperience", y="Salary", data=df)
plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

# -----------------------------------------
# 13. Salary by Department
# -----------------------------------------

plt.figure()
sns.boxplot(x="Department", y="Salary", data=df)
plt.title("Salary by Department")
plt.xticks(rotation=30)
plt.show()

# -----------------------------------------
# 14. Salary vs Performance
# -----------------------------------------

plt.figure()
sns.boxplot(x="Performance", y="Salary", data=df)
plt.title("Salary vs Performance")
plt.show()

# -----------------------------------------
# 15. Correlation Analysis
# -----------------------------------------

correlation = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------------------
# 16. Key Insights
# -----------------------------------------

print("\n==============================")
print("Basic Insights from the Data")
print("==============================")

print("1. Salary tends to increase with years of experience.")
print("2. Some departments have higher salary ranges than others.")
print("3. Most employees fall within the mid-age range.")
print("4. Performance rating may influence salary levels.")

print("\nEDA Completed Successfully!")
```
